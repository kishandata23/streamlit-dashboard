```python
plt.figure(figsize=(5,5))
gender_dist = sns.countplot(df,x='Gender')
gender_dist.bar_label(gender_dist.containers[0])
plt.show()
```

```python
group_by_ParentEduc = df.groupby('ParentEduc').agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(group_by_ParentEduc)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(group_by_ParentEduc,annot=True)
plt.show()
```

Higher educated parent's childrens scrore in subjec are Very good
